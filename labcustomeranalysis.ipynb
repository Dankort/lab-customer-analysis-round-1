```python

df3 = pd.read_csv("/Users/danort/Desktop/Ironhack labs/lab-customer-analysis-round-1/files_for_lab/csv_files/file3.csv")
df3.head()
df3.shape
df3.columns = df3.columns.str.upper().str.replace(' ', '_').str.replace('[^a-zA-Z0-9_]', '')

#changing the name of a column

df3 = df3.rename(columns={'STATE' : 'ST'})
df3







```

```python
#Rearranging
df1 = df1[["CUSTOMER", "ST", "GENDER", "EDUCATION", "INCOME", "CUSTOMER_LIFETIME_VALUE", "NUMBER_OF_OPEN_COMPLAINTS", "VEHICLE_CLASS", "POLICY_TYPE", "MONTHLY_PREMIUM_AUTO", "TOTAL_CLAIM_AMOUNT"]]
df2 = df2[["CUSTOMER", "ST", "GENDER", "EDUCATION", "INCOME", "CUSTOMER_LIFETIME_VALUE", "NUMBER_OF_OPEN_COMPLAINTS", "VEHICLE_CLASS", "POLICY_TYPE", "MONTHLY_PREMIUM_AUTO", "TOTAL_CLAIM_AMOUNT"]]
df3 = df3[["CUSTOMER", "ST", "GENDER", "EDUCATION", "INCOME", "CUSTOMER_LIFETIME_VALUE", "NUMBER_OF_OPEN_COMPLAINTS", "VEHICLE_CLASS", "POLICY_TYPE", "MONTHLY_PREMIUM_AUTO", "TOTAL_CLAIM_AMOUNT"]]

display(df1)
display(df2)
display(df3)
```

```python
data = pd.concat([df1,df2,df3], axis=0).reset_index(drop=True)
data
```

```python
#Which columns are numerical

numerical_columns = data.select_dtypes(include=['int', 'float']).columns
numerical_columns
```

```python
#Which columns are categorical
categorical_columns = data.select_dtypes(include=['category']).columns
categorical_columns
```

```python
#Delete the column education and the number of open complaints from the dataframe.

data.drop(['EDUCATION', 'NUMBER_OF_OPEN_COMPLAINTS'], axis=1, inplace=True)
```

```python
#Correct the values in the column customer lifetime value. They are given as a percent, so multiply them by 100 and change dtype to numerical type

data['CUSTOMER_LIFETIME_VALUE'] = data['CUSTOMER_LIFETIME_VALUE'].astype(str).str.rstrip('%').astype('float')*100.0
data
```

```python
#Check for duplicate rows in the data and remove if any.

data

duplicate_rows = data[data.duplicated()]
duplicate_rows
```

```python
#Remove duplicate

data.drop_duplicates(inplace=True)
data
```
